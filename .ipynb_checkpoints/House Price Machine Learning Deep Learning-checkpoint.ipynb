## Splitting the dataset

```python
pre_processed_df = housing_df_v2
pre_processed_df[:10]
```

```python
y = pre_processed_df['Price']
X = pre_processed_df.drop(['Price'], axis = 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
X.info()
```

```python
X_test[:50]
```

```python
y[:50]
```

```python
# # Preprocess numerical data for neural network

# # Create a StandardScaler instances
# scaler = StandardScaler()

# # Fit the StandardScaler
# X_scaler = scaler.fit(X_train)

# # Scale the data
# X_train_scaled = X_scaler.transform(X_train)
# X_test_scaled = X_scaler.transform(X_test)
```

```python
# Preprocess numerical data for neural network

# Create a StandardScaler instances
scaler = MinMaxScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
X_test.shape
```

```python
model=Sequential()

model.add(Dense(20,activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dense(20,activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(X_train,y_train, validation_data=(X_test,y_test), batch_size=150, epochs=250 )
```

```python
loss_df=pd.DataFrame(model.history.history)
```

```python
loss_df.plot(figsize=(7,6), colormap='RdBu_r')
plt.title('Training Loss Per Epoch', fontsize=15)
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Loss', fontsize=12)
plt.show()
```

```python
model.evaluate(X_test,y_test,verbose=0)
```

```python
model.evaluate(X_train,y_train,verbose=0)
```

```python
prediction=model.predict(X_test)
```

```python
plt.figure(figsize=(8,6))
plt.plot(y_test,y_test, color='b')
plt.scatter(y_test,prediction,color='r')
plt.xlabel('Actual Price', fontsize=12)
plt.ylabel('Predicted Price', fontsize=12)
plt.title('Predicted Versus Actual Prices', fontsize=15)
plt.show()
```

```python
print('MAE= ', metrics.mean_absolute_error(y_test,prediction))
print('MSE= ', metrics.mean_squared_error(y_test,prediction))
print('RMSE= ', np.sqrt(metrics.mean_squared_error(y_test,prediction)))
print('R2 Score= ', metrics.r2_score(y_test,prediction))
```
