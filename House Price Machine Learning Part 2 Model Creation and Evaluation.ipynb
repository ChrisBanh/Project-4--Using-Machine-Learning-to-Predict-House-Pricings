## Part 2: Training the model and evaluation w/ optimisations 

### The information was directly pulled from the "housing" SQL database and converted into a pandas dataframe prior to evaluation. 

```python
# Import our dependencies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import r2_score
import numpy as np

```

## Splitting the dataset

```python
from sqlalchemy import create_engine, func
from  config import password 
```

```python
# Create the engine that will allow to pull the table from the PostgresSQL

engine = create_engine(f'postgresql://postgres:{password}@localhost:5432/housing_db')
```

```python
# Read the SQL Table into a dataframe

pd.read_sql("SELECT * FROM housing", engine)
```

```python
#Ensure that the table is identical prior to been processed within PostgresSQL

pre_trained_df = pd.read_sql("SELECT * FROM housing", engine)
pre_trained_df.info()
```

```python
pre_trained_df = pre_trained_df.drop(columns=['id'])
pre_trained_df.info()
```

```python
# Split the dataset into X and y


y = pre_trained_df['price']
X = pre_trained_df.drop(['price'], axis = 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
X.info()
```

```python
X_test[:50]
```

```python
y[:50]
```

```python
# Initiate the model using RandomForestRegressor


model = RandomForestRegressor(n_estimators=18, random_state=500)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
# Get the MAE, MSE and RMSE


mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

```python
# Get the R^2 value. The threshold for models like these is at least a 0.8 for it to be deemed sufficiently accurate 

r2 = r2_score(y_test, y_pred)

print(f"R-squared (R^2) score: {r2}")
```

```python
model = RandomForestRegressor(n_estimators=40, random_state=500)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

### Manual optimisation attempt 3: Change the Hyperparameters around

Adjusting the n_parameter from 50 to 55. Also included the max_features hyperparameter which is set as 8.

```python
model = RandomForestRegressor(n_estimators=55, random_state=500, max_features = 8)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")

r2 = r2_score(y_test, y_pred)


print(f"R-squared (R^2) score: {r2}")
```

#### Even though we have already achieved the minimum threshold R62 value of 0.80, we believe we can still achieve a higher value, this time creating a script that uses GridSearchCV to determine the best hyperparameters from a parameters list. 


```python
# Initialising a script that determines the best hyper parameters. 

# Define the hyperparameters and their possible values
param_grid = {
    'n_estimators': [40, 55, 70, 100, 200, 300, 400],
    "max_features" : [2, 4, 6, 8, 10, 12, 14, 20],
    'random_state': [42, 400, 500, 600, 800, 1000]  # Here we're adding three different seeds as an example
}

# Initialize grid search
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='neg_mean_squared_error', return_train_score = True, n_jobs=-1)


# Fit the model
grid_search.fit(X_train, y_train)

# Evaluate on test data
best_rf = grid_search.best_estimator_
y_pred = best_rf.predict(X_test)
print(best_rf)
```

```python
# Finally obtain the MAE, MSE, RMSE and the R62 value using the best parameters obtained through the GridSearch tuning. 


mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")

r2 = r2_score(y_test, y_pred)
print(f"R-squared (R^2) score: {r2}")
```

```python
# Compare the difference in values between the predicted and the test values for price. 

diff = y_test - y_pred
plt.plot(diff, 'rd')
plt.ylabel('Absolute Price Error ($)')
plt.show()
```
